# K-means cluster (Elbow method)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Sample student heights in cm
heights = np.array([150, 152, 154, 157, 160, 165, 170, 175, 180, 183, 185, 188, 190]).reshape(-1, 1)
```

```python
# Elbow Method: Try different values of K
wcss = []  # To store Within-Cluster Sum of Squares

for k in range(1, 6):  # Trying K = 1 to 5
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(heights)
    wcss.append(kmeans.inertia_)  # .inertia_ gives WCSS
```

```python
k = [i for i in range(1,6)]
```

```python
# Plot the Elbow Curve
plt.plot(range(1, 6), wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Inertia)')
plt.grid(True)
plt.show()


# I have a model with 5 K values
# k = 1    --> wcss
# k = 2    --> wcss
# k = 3    --> wcss
# k = 4    --> wcss
# k = 5    --> wcss
```
