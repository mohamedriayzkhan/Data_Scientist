# K - Means Cluster Algorithm (Unsupervised)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Step 1: Sample student heights (in cm)
heights = [150, 152, 154, 157, 160, 162, 165, 168, 170, 173, 175, 178, 180, 183, 185, 188, 190]
print("Sample heights:", heights)
```

```python
print(heights)
```

```python
heights = np.array(heights).reshape(-1, 1)  # reshape for sklearn   #reshape(rows, columns)
print("Reshaped heights for clustering:", heights)

# -1 - figure out this dimension automatically based on the size of the data and the other dimensions
```

```python
# Step 2: Apply K-Means (let’s choose 3 clusters: short, medium, tall)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(heights)
```

```python
# Step 3: Get cluster labels
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
print("Cluster labels for each height:", labels)
print("Centroids of clusters:", centroids)
```

```python
# Step 4: Plot the clusters
plt.figure(figsize=(10, 4))
plt.scatter(heights, np.zeros_like(heights), c=labels, cmap='viridis', s=100)   #(150,0)
plt.scatter(centroids, [0,0,0] , c='red', s=200, marker='X', label='Centroids') #[0,0,0]
plt.title('K-Means Clustering of Student Heights')
plt.xlabel('Height (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

plt.scatter(heights, np.zeros_like(heights), c=labels, cmap='viridis', s=100)

| Part                     | Meaning                                                                      |
| ------------------------ | ---------------------------------------------------------------------------- |
| `heights`                | X-axis values (student heights in cm)                                        |
| `np.zeros_like(heights)` | Y-axis values — all **zeros** so that we see points **on a horizontal line** |
| `c=labels`               | Colors are based on the **cluster labels** (0, 1, 2)                         |
| `cmap='viridis'`         | A **colormap** to automatically assign different colors to clusters          |
| `s=100`                  | Size of each point (bubble size = 100)                                       |


plt.scatter(centroids, [0, 0, 0], c='red', s=200, marker='X', label='Centroids')

| Part                | Meaning                                                          |
| ------------------- | ---------------------------------------------------------------- |
| `centroids`         | X-axis positions of the cluster centers (means of each group)    |
| `[0, 0, 0]`         | Y-axis values — all zeros to line them up horizontally           |
| `c='red'`           | Color of centroid markers is **red**                             |
| `s=200`             | Size is **bigger** than normal points so they stand out          |
| `marker='X'`        | Shape of the centroid is an **'X'**                              |
| `label='Centroids'` | Adds a label for the **legend** (so we know these are centroids) |
