```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
heights = [150, 160, 165, 170, 175, 180, 185, 190, 195, 200, 205, 210, 215, 220, 225, 230, 235, 240, 245, 250]
print(heights)
```

```python
heights = np.array(heights).reshape(-1, 1)
print(heights)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(heights)
```

```python
lables = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
print("Labels:", lables)
print("Centroids:", centroids)
```

```python
# Step 4: Plot the clusters
plt.figure(figsize=(10, 4))
plt.scatter(heights, np.zeros_like(heights), c=lables, cmap='viridis', s=100)   #(150,0)
plt.scatter(centroids, [0,0,0] , c='red', s=200, marker='X', label='Centroids') #[0,0,0]
plt.title('K-Means Clustering of Student Heights')
plt.xlabel('Height (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
for i, h in enumerate(heights):
    print(f"Height: {h[0]} cm, Cluster: {lables[i]}")
```
