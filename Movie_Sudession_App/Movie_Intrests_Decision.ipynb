```python
import pandas as pd
movie_info = pd.read_csv('/home/intellect/Documents/Data_Scientist/DecisionTree_Classifier/Movie_Interests_DecisionTree.csv')
movie_info.head()
```

```python
input_dataset = movie_info.drop('Interest', axis=1)
input_dataset.head()
```

```python
out_dataset = movie_info['Interest']
out_dataset.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(input_dataset, out_dataset)
movie_intrest  = model.predict([[14,0],[40,1]])
movie_intrest
```
